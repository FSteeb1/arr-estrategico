```python
import pandas as pd
import numpy as np

# 1. CARGA DE DATOS
df_resultados = pd.read_csv('C:/Users/felip/Documents/Cosas de Peter/4n4lisis de d4tos/Resultados_limpios.csv') 
df_demograficos = pd.read_csv('C:/Users/felip/Documents/Cosas de Peter/4n4lisis de d4tos/Dem_limpios.csv')
```

```python
# --- PASO 1: LIMPIEZA (Excluir a los que fallaron el Attention Check) ---

# hecho en sheets

# --- PASO 2: ESTADÍSTICOS DESCRIPTIVOS (Nivel Respuesta) ---

# (Asumiendo que solo te interesan las respuestas a los dilemas reales)
df_analisis = df_resultados[~df_resultados['dilema'].str.contains('Atencion', case=False, na=False)]

print("\n--- Frecuencia de Respuestas (Total Trials) ---")
conteo_respuestas = df_analisis['respuesta'].value_counts()
porcentaje_respuestas = df_analisis['respuesta'].value_counts(normalize=True) * 100

resumen_respuestas = pd.DataFrame({
    'Frecuencia': conteo_respuestas,
    'Porcentaje': porcentaje_respuestas.round(2)
})

print(resumen_respuestas)
```

```python
# --- PASO 3: DESCRIPCIÓN DE LA MUESTRA (Nivel Sujeto) ---

print("\n--- Demográficos de la Población Final ---")

# Género (Asumiendo columna 'Genero')
genero_counts = df_demograficos['Genero'].value_counts()
genero_pct = df_demograficos['Genero'].value_counts(normalize=True) * 100

resumen_genero = pd.DataFrame({
    'N': genero_counts,
    'Porcentaje': genero_pct.round(2)
})
print("\nDistribución por Género:")
print(resumen_genero)
```
